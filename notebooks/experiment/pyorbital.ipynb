```python
df['solarazang'] = df.apply(lambda x: np.degrees(astronomy.get_alt_az(x['date'], x['lon'], x['lat'])[1]),axis=1)
```

```python
df['solarzenang'] = df.apply(lambda x: astronomy.sun_zenith_angle(x['date'], x['lon'], x['lat']), axis=1)
```

```python
df['relazang'] = df['solarazang'] - df['viewazang']
```

```python
df_eq = df[(df['lat'] >= -0.5 ) & (df['lat'] <= 0.5 )]
```

```python
df.head()
df['hour'] = df['date'].dt.hour
df.plot.scatter(x='hour', y='solarzenang')
```

```python
def compute_sun_glint_cos(row):
    theta_v = np.radians(row['viewzenang'])
    theta_s = np.radians(row['solarzenang'])
    phi = np.radians(row['relazang'])
    cos_sunglint = np.cos(theta_v) * np.cos(theta_s) - (np.sin(theta_v) * np.sin(theta_s) * np.cos(phi) )
    return cos_sunglint
```

```python
df['cos_sunglint'] = df.apply(compute_sun_glint_cos, axis=1)
```

```python
df['sunglint_angle'] = np.degrees(np.arccos(df['cos_sunglint']))
df['sunglint_angle'].hist()
```

$$ cos \theta_g = cos \theta_v cos \theta_s - sin \theta_v sin \theta_s cos \phi$$ 

where 

$\theta_g$ = sun glint angle

$\theta_v$ = satellite view/zenith angle

$\theta_s$ = solar zenith angle

$\phi$ = relative azimuth angle (difference between solar and satellite azimuth angle)
